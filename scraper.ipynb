Web Scraper Project Using Beautiful Soup


```python
#import libraries to use
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/Top_100_Contractors_of_the_U.S._federal_government'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find('table')
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_= 'wikitable sortable')
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
top_titles = table.find_all('th')
```

```python
top_titles
```

```python
top_table_titles = [title.text.strip() for title in top_titles]

print(top_table_titles)
```

```python
top_table_titles.remove(top_table_titles[0])
```

```python
df = pd.DataFrame(columns = top_table_titles)
df
```

```python
column_data = table.find_all('tr')
column_data.remove(column_data[0])
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
df['Rank']= df['Global vendor name'].str.split('.').str[0]
df['Global vendor name'] = df['Global vendor name'].str.slice(4)
```

```python
df
```

```python
cols = list(df.columns.values)
cols
```

```python
df = df[['Rank','Global vendor name',
 'Dollars obligated (millions)',
 '%Total dollars']]
df
```

```python
df.to_csv(r'GovtContractors2019.csv', index = False)
```
